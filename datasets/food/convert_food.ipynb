```python
import pandas as pd
import os 
```

**Bundle Item**

```python
def create_mapping_item(item_id):
    item_id = sorted(set(item_id))
    item_id_mapping = dict()
    for i in range(len(item_id)):
        item_id_mapping[item_id[i]] = i
    return item_id_mapping
```

```python
df_bi = pd.read_csv('./bundle_item.csv')
df_bi.head()
```

```python
item_id = []
for i in range(df_bi.shape[0]):
    item_id.append(df_bi.iloc[i]['item ID'])
mapping_item_id = create_mapping_item(item_id)
print(mapping_item_id)
```

```python
df_bi['item ID'] = df_bi['item ID'].map(mapping_item_id)
df_bi.head()
```

```python
path = './bundle_item.txt'
if os.path.exists(path):
    os.remove(path)

bundle = []
item = []

with open(path, 'a') as f:
    for i in range(df_bi.shape[0]):
        row = df_bi.iloc[i]
        x, y = row['bundle ID'], row['item ID']
        bundle.append(x)
        item.append(y)
        f.write(f'{x}\t{y}\n')

num_bundle = len(set(bundle))
num_item = len(set(item))
```

**User Item**

```python
df_ui = pd.read_csv('./user_item.csv')
df_ui['item ID'] = df_ui['item ID'].map(mapping_item_id)
df_ui = df_ui.drop(columns=[
    'timestamp'
])
df_ui.head()
```

```python
path = './user_item.txt'
if os.path.exists(path):
    os.remove(path)

with open(path, 'a') as f:
    for i in range(df_ui.shape[0]):
        row = df_ui.iloc[i]
        x, y = row['user ID'], row['item ID']
        f.write(f'{x}\t{y}\n')
```

**User bundle**

```python
df_ub = pd.read_csv('./user_bundle.csv')
df_ub = df_ub.drop(columns=['timestamp'])
df_ub.head()
```

```python
user_id = []
bundle_id = []

for i in range(df_ub.shape[0]):
    row = df_ub.iloc[i]
    user_id.append(row['user ID'])
    bundle_id.append(row['bundle ID'])

from sklearn.model_selection import train_test_split
x_train, x_test_val, y_train, y_test_val = train_test_split(user_id, bundle_id, test_size=0.2)
x_test, x_val, y_test, y_val = train_test_split(x_test_val, y_test_val, test_size=0.5)

name_task = ['train', 'test', 'tune']
for task in name_task:
    path = f'user_bundle_{task}.txt' 
    if os.path.exists(path):
        os.remove(path)
    with open(path, 'a') as f:
        if task == 'train':
            for i in range(len(x_train)):
                f.write(f'{x_train[i]}\t{y_train[i]}\n')
        elif task == 'tune':
            for i in range(len(x_val)):
                f.write(f'{x_val[i]}\t{y_val[i]}\n')
        else:
            for i in range(len(x_test)):
                f.write(f'{x_test[i]}\t{y_test[i]}\n')                        
```

```python
num_user = len(set(user_id))
```
